```python
import numpy as np;
import pandas as pd;
import seaborn as sns;
import matplotlib.pyplot as plt;
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df,x ="Gender" )
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show();
```

```python
#From the above chart we obserbed that the number of male is greater than number of female
```

```python
gb   = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore" :"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot = True)
plt.title("Relationship between Parent's Education and student'score")
plt.show()
```

```python
gb1   = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore" :"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot = True)
plt.title("Relationship between Parent's Marital Status and student'score")
plt.show()
```

```python
#from the above cgart we have concluded that there is negligible impact on the studenet's score due to their parent's marital status
```

```python
gb2   = df.groupby("WklyStudyHours").agg({"MathScore":"mean","ReadingScore" :"mean","WritingScore":"mean"})
print(gb2)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb2,annot = True)
plt.title("Relationship between Wkly Study Hours and student'score")
plt.show()
```

Distribution of Ethnic Groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A","group B","group C","group D","group E"]

mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist,labels =l,autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()


```
